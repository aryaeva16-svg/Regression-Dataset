```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from google.colab import files
uploaded = files.upload()
```

```python
customer_df = pd.read_csv("Customer_Purchase_Dataset (updated ver.).csv")

customer_df.head()
customer_df.info()
```

```python
customer_df["HighSpender"] = (
    (customer_df["Spending_Score"] > customer_df["Spending_Score"].median()) &
    (customer_df["Purchase_Frequency"] > customer_df["Purchase_Frequency"].median())
).astype(int)

customer_df["HighSpender"].value_counts()
X = customer_df[
    ["Age",
     "Income",
     "Membership_Years",
     "Purchase_Frequency",
     "Last_Purchase_Amount"]
]

y = customer_df["HighSpender"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_prob = model.predict_proba(X_test_scaled)[:, 1]
def classify(probabilities, threshold):
    return (probabilities >= threshold).astype(int)
y_pred_05 = classify(y_prob, 0.5)

print("Threshold = 0.5")
print(confusion_matrix(y_test, y_pred_05))
print(classification_report(y_test, y_pred_05))
```

```python
y_pred_07 = classify(y_prob, 0.7)

print("Threshold = 0.7")
print(confusion_matrix(y_test, y_pred_07))
print(classification_report(y_test, y_pred_07))
```

```python
y_pred_04 = classify(y_prob, 0.4)

print("Threshold = 0.4")
print(confusion_matrix(y_test, y_pred_04))
print(classification_report(y_test, y_pred_04))
```

```python
coef_df = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_[0]
})

coef_df
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_05)
cm
```
